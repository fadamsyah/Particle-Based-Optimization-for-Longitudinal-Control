```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
plt.style.use(['science','ieee']) # Ini kudu install install dulu
```

```python
# Data lama

gt = {}
v_max = -np.Inf
v_min = np.Inf
t_max = -np.Inf
t_min = np.Inf
for i in range(1, 13):
    gt[i] = {}
    gt[i]['t'] = np.array(pd.read_csv(str(i-1) + '.csv')['t'].values)
    gt[i]['v'] = np.array(pd.read_csv(str(i-1) + '.csv')['v'].values)
    if np.max(gt[i]['v']) >= v_max:
        v_max = np.copy(np.max(gt[i]['v']))
    if np.min(gt[i]['v']) <= v_min:
        v_min = np.copy(np.min(gt[i]['v']))
    if np.max(gt[i]['t']) >= t_max:
        t_max = np.copy(np.max(gt[i]['t']))
    if np.min(gt[i]['t']) <= t_min:
        t_min = np.copy(np.min(gt[i]['t']))
        
label = ['1st', '2nd', '3rd', '4th',
         '5th', '6th', '7th', '8th',
         '9th', '10th', '11th', '12th']

for i in range(3):
    plt.clf()
    for j in range(1, 5):
        idx = 4*i + j
        plt.plot(gt[idx]['t'], gt[idx]['v'], label=label[idx - 1])
    plt.legend()
    plt.title('Ground Truth Data')
    plt.xlabel('Time (s)')
    plt.ylabel('Speed (m/s)')
    plt.xlim(int(t_min), int(t_max + 1.5))
    plt.ylim(int(v_min), int(v_max + 1.5))
    plt.savefig('img/ground_truth_lama_' + str(i + 1), dpi=600, transparent=True, bbox_inches='tight')
```

```python
# Step Response

gt = {}
v_max = -np.Inf
v_min = np.Inf
t_max = -np.Inf
t_min = np.Inf
for i in range(13, 25):
    gt[i] = {}
    gt[i]['t'] = np.array(pd.read_csv('s' + str(i-13) + '.csv')['t'].values)
    gt[i]['v'] = np.array(pd.read_csv('s' + str(i-13) + '.csv')['v'].values)
    if np.max(gt[i]['v']) >= v_max:
        v_max = np.copy(np.max(gt[i]['v']))
    if np.min(gt[i]['v']) <= v_min:
        v_min = np.copy(np.min(gt[i]['v']))
    if np.max(gt[i]['t']) >= t_max:
        t_max = np.copy(np.max(gt[i]['t']))
    if np.min(gt[i]['t']) <= t_min:
        t_min = np.copy(np.min(gt[i]['t']))
        
label = ['1st', '2nd', '3rd',
         '4th', '5th', '6th',
         '7th', '8th', '9th',
         '10th', '11th', '12th',
         '13th', '14th']

for i in range(3):
    plt.clf()
    for j in range(4):
        idx = 4*i + j + 13
        if idx <= len(gt) + 13:
            plt.plot(gt[idx]['t'], gt[idx]['v'], label=label[idx - 13])
    plt.legend()
    plt.title('Ground Truth Data')
    plt.xlabel('Time (s)')
    plt.ylabel('Speed (m/s)')
    plt.xlim(int(t_min), int(t_max + 1.5))
    plt.ylim(int(v_min), int(v_max + 1.5))
    plt.savefig('img/ground_truth_step_' + str(i + 1), dpi=600, transparent=True, bbox_inches='tight')
```
