```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from copy import deepcopy
from os import listdir
from os.path import isfile, join

plt.style.use(['science', 'grid'])

DPI = 500
```

# Load Variables

```python
df = pd.read_csv('carla_sim.csv')
t = df['time'].values
sp = df['sp_v'].values
gt = df['speed'].values
throttle = df['throttle'].values
brake = df['brake'].values
```

```python
plt.plot(t, gt)
plt.plot(t, sp)
```

# PLOT

```python
def get_array_from_time_interval(x, t, T):
    t0 = min(T)
    t1 = max(T)
    i1 = -1
    i0 = 0
    for i in range(len(t)):
        if t0 <= t[i]:
            i0 = deepcopy(i)
            break
    for i in range(i0, len(t)):
        if t1 <= t[i]:
            i1 = deepcopy(i)
            break

    return x[i0:i1], t[i0:i1]
```

## Plot 1

```python
plt.plot(t, sp, label='Setpoint', zorder=2, color='black', linestyle='--')
plt.plot(t, gt, label='Response', zorder=1, color='red', linewidth=2.)

plt.xlabel('Time $(s)$')
plt.ylabel('Speed $(m/s)$')
plt.legend(fontsize=7)
plt.savefig(f'img/sim_result.png', dpi=DPI, transparent=True, bbox_inches='tight')
plt.show()
```

```python
plt.plot(t, throttle, color='black', linewidth=0.5)

plt.xlabel('Time $(s)$')
plt.ylabel('Throttle')
plt.ylim([0,None])
plt.savefig(f'img/sim_result_throttle.png', dpi=DPI, transparent=True, bbox_inches='tight')
plt.show()
```

## Plot 2

```python
T = [0, 4.5]
NAME = 'r1'
l1, lt1 = get_array_from_time_interval(gt, t, T)
l2, lt2 = get_array_from_time_interval(sp, t, T)
l3, lt3 = get_array_from_time_interval(throttle, t, T)
```

```python
plt.plot(lt2, l2, label='Setpoint', zorder=2, color='black', linestyle='--')
plt.plot(lt1, l1, label='Response', zorder=1, color='red', linewidth=2.)

plt.xlabel('Time $(s)$')
plt.ylabel('Speed $(m/s)$')
plt.legend(fontsize=7)
plt.savefig(f'img/{NAME}.png', dpi=DPI, transparent=True, bbox_inches='tight')
plt.show()
```

```python
plt.plot(lt3, l3, color='black', linewidth=1.25)

plt.xlabel('Time $(s)$')
plt.ylabel('Throttle')
plt.ylim([0,None])
plt.savefig(f'img/{NAME}_throttle.png', dpi=DPI, transparent=True, bbox_inches='tight')
plt.show()
```
